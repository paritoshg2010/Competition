```python
from datetime import datetime

#function to calculate difference between time
def time_diff (t1,t2):
    FMT = '%H:%M:%S'
    tdelta = (datetime.strptime(t2, FMT) - datetime.strptime(t1, FMT)).seconds
    return tdelta
```

```python
#populating time taken in each trajectory
data ['time_taken'] = np.vectorize(time_diff)(data['time_entry'],data['time_exit'])
```

```python
#dropping the first column as its redundant as index
data.drop(data.columns[0], axis=1, inplace=True)

#function to calculate distance between two co-ordinates
def disp (x1, y1, x2, y2):
    
    disp = (((x1-x2)**2) + ((y1-y2)**2))**0.5
    return disp
    
```

```python
#populating distance between the co-ordinate
data['disp'] = np.vectorize(disp)(data['x_entry'], data['y_entry'], data['x_exit'], data['y_exit'])
```

```python
#function to calculate velocity for the trajectory
def velocity(disp, time):
    
    if time != 0:
        vel = disp/time
    else:
        vel = 0
    return vel
```

```python
#populating velocity for the trajectory
data['vel'] = np.vectorize(velocity)(data['disp'], data['time_taken'])
```

```python
#city center co-ordinates
xc1 = 3750901.506
xc2 = 3770901.5068
yc1 = -19268905.6133
yc2 = -19208905.6133 


#function to populate the target coulmn whose exit co-ordinate in city center between 15:00-16:00
def target(t2, x2, y2):
    target = float("NaN")
    if (t2 >= '15:00:00') and (t2 <= '16:00:00'):
        if (x2>=xc1) and (x2<=xc2) and (y2>=yc1) and (y2<=yc2):
            target = int(1)
        else:
            target = int(0)
    
    return target
    
```

```python
#populating the target column
data['target'] = np.vectorize(target)(data['time_exit'], data['x_exit'], data['y_exit'])
```

```python
#city center co-ordinates
xc1 = 3750901.506
xc2 = 3770901.5068
yc1 = -19268905.6133
yc2 = -19208905.6133 


#function to populate the target coulmn whose exit co-ordinate in city center 
def target_all(x2, y2):
    target = float("NaN")
    if (x2>=xc1) and (x2<=xc2) and (y2>=yc1) and (y2<=yc2):
        target = int(1)
    else:
        target = int(0)
    
    return target
```

```python
data['target_all'] = np.vectorize(target_all)(data['x_exit'], data['y_exit'])
```

Below function will populate the probability of last trajectory for each hash id's

```python
from tqdm import tqdm_notebook as tqdm

#creating probability, mean_disp, mean_time column with NA values
data['prob']= float('NaN')
data['mean_disp'] = float('NaN')
data['mean_time'] = float('NaN')


#populating the values
for hash_val in tqdm(df_hash):
    
    mean_disp = data[data.hash == hash_val]['disp'][:-1].mean()
    mean_time = data[data.hash == hash_val]['time_taken'][:-1].mean()
    prob = data[data.hash == hash_val]['target_all'][:-1].mean()
    idx = data[data['hash'] == hash_val][-1:].index[0]
    data.iloc[idx, data.columns.get_loc('mean_disp')] = mean_disp
    data.iloc[idx, data.columns.get_loc('mean_time')] = mean_time
    data.iloc[idx, data.columns.get_loc('prob')] = prob
```

```python
data['entry_tar_all'] = np.vectorize(target_all)(data.x_entry, data.y_entry)
data['entry_prob'] = float(0)
```

```python
for hash_val in tqdm(df_hash):
    entry_prob = data[data.hash == hash_val]['entry_tar_all'][:-1].mean()
    idx = data[data['hash'] == hash_val][-1:].index[0]
    data.iloc[idx, data.columns.get_loc('entry_prob')] = entry_prob
    
```

```python
from math import sin, cos, sqrt, atan2, radians

R = 6373.0
xc_cent= (xc1+xc2)/2
yc_cent= (yc1+yc2)/2

def dist_center(x1,y1):
    
    lat1 = radians(y1)
    lon1 = radians(x1)
    lat2 = radians(yc_cent)
    lon2 = radians(xc_cent)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
